```python
import re

# 解析 score 计算 胜负盘数、总局数、局差
def parse_score_extended(score):
    if pd.isna(score):
        return None, None, None, None, None, None  # 处理缺失值
    
    # 处理退赛情况
    ret_flag = 1 if "RET" in score else 0
    score = score.replace("RET", "").strip()  # 移除 RET，防止干扰解析
    
    # 拆分比分
    score_parts = str(score).split(" ")

    win_sets, lose_sets = 0, 0  # 统计胜负盘数
    total_win_games, total_lose_games = 0, 0  # 统计胜负总局数

    for part in score_parts:
        match = re.match(r"(\d+)-(\d+)(?:\((\d+)\))?", part)
        if match:
            p1, p2 = int(match.group(1)), int(match.group(2))
            tiebreak = int(match.group(3)) if match.group(3) else None
            
            # 处理抢七局，如果抢七分数 > 4，视为 8-6
            if tiebreak and tiebreak > 4:
                p1, p2 = 8, 6

            # 统计胜负盘数
            if p1 > p2:
                win_sets += 1
            else:
                lose_sets += 1

            # 统计总局数
            total_win_games += p1
            total_lose_games += p2

    # 计算局差
    game_difference = total_win_games - total_lose_games

    return win_sets, lose_sets, total_win_games, total_lose_games, game_difference, ret_flag
```

```python
# 遍历所有 DataFrame 并应用解析函数
for name, df in dfs.items():
    df[["win_sets", "lose_sets", "total_win_games", "total_lose_games", "game_diff", "ret"]] = df["score"].apply(
        lambda s: pd.Series(parse_score_extended(s))
    )
```

```python
import importlib
import welo
from welo import WEloRatingSystem
importlib.reload(welo)
```

```python
# 初始化 Elo 计算器
elo_system = WEloRatingSystem()

# 确保 df 按 `tourney_date` 排序
for key, df in dfs.items():
    #df["tourney_date"] = pd.to_datetime(df["tourney_date"], format="%Y%m%d")
    df = df.sort_values(by="tourney_date").reset_index(drop=True)
    
    # 计算 Elo 并存储预测胜率
    expected_scores = []
    for _, row in df.iterrows():
        winner_id, loser_id = row["winner_id"], row["loser_id"]
        winner_games, loser_games = row["total_win_games"], row["total_lose_games"]
        #存储比赛前elo排名
        #df.loc["winner_elo_before"] = df["winner_id"].map(elo_system.ratings)
        df["winner_elo_before"] = df["winner_id"].apply(lambda x: elo_system.ratings.get(x, 1500))
        df["loser_elo_before"] = df["loser_id"].apply(lambda x: elo_system.ratings.get(x, 1500))

        # 计算预期胜率
        expected = elo_system.predict_outcome(winner_id, loser_id)
        expected_scores.append(expected)

        # 更新 Elo 评分
        elo_system.update_rating(winner_id, loser_id, outcome=1, winner_game=winner_games, loser_game=loser_games)

        # 将 Elo 评分和预测胜率加入 df
        # df["winner_elo_after"] = df["winner_id"].map(elo_system.ratings)
        # df["loser_elo_after"] = df["loser_id"].map(elo_system.ratings)
    df["expected_score"] = expected_scores

    # 计算实际比赛结果 vs 预测胜率的差距
    df["elo_prediction_error"] = 1 - df["expected_score"]
    
    dfs[key] = df

    #print(df[["tourney_date", "winner_id", "winner_elo", "loser_id", "loser_elo", "expected_score", "elo_prediction_error"]])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))  # 设置画布大小

for key, df in dfs.items():
    # 绘制每个数据集的分布
    sns.histplot(df['elo_prediction_error'], bins=30, kde=True, label=key, alpha=0.6)

# 图例与标签
plt.xlabel("elo_prediction_error")
plt.ylabel("Frequency")
plt.title("Distribution of elo_prediction_error across datasets")
plt.legend()  # 显示不同数据集的图例
plt.grid(True)
plt.show()
```

```python
# 创建统一画布
plt.figure(figsize=(12, 6))

# 循环遍历每个数据集
for key, df in dfs.items():
    # 打印统计信息（保留原始功能）
    print(df[['elo_prediction_error']].describe())
    
    # 在同一坐标系绘制直方图
    sns.histplot(
        data=df,
        x='elo_prediction_error',
        bins=30,
        kde=True,
        alpha=0.4,  # 设置透明度防止重叠遮挡
        label=key    # 使用键名作为图例标签
    )

# 统一设置图表元素
plt.xlabel('elo_prediction_error')
plt.ylabel('Density' if plt.gca().get_ylabel() == 'Count' else 'Frequency')  # 自适应y轴标签
plt.title('Comparative Distribution of elo_prediction_error')
plt.legend(title='Dataset')  # 添加分类图例
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()

# 展示图表
plt.show()
```

```python
# 筛选高误差数据
high_error_data = {key: df[df["elo_prediction_error"] > 0.9] for key, df in dfs.items()}

# 分析Elo差异分布
for key, data in high_error_data.items():
    if not data.empty:
        data["elo_diff"] = data["loser_elo_before"] - data["winner_elo_before"]
        print(f"数据集 {key}:")
        print(f"  - 平均Elo差异: {data['elo_diff'].mean()}")
        print(f"  - 最大Elo差异: {data['elo_diff'].max()}")
        print(f"  - 新选手比例: {(data['winner_elo_before'] < 1600).mean() * 100:.1f}%")

# 检查比分悬殊情况
high_error_df = pd.concat(high_error_data.values())
high_error_df["score_ratio"] = high_error_df["total_win_games"] / high_error_df["total_lose_games"].replace(0, 1e-6)
print(f"比分悬殊比赛占比（ratio > 2）: {(high_error_df['score_ratio'] > 2).mean() * 100:.1f}%")
```
